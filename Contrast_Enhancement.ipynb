```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
def plot_image(image_1, image_2, title_1="Original", title_2="New Image"):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
```

```python
image_BGR = cv2.imread("mycat.png")
image_RGB = cv2.cvtColor(image_BGR, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (8,8))
plt.imshow(image_RGB)

plt.show()
```

```python
def powerLowTrans(input_img, l, c):
    
    norm_img = input_img / 255
    image_out = c*norm_img**(l)
    denorm_img = np.round(image_out*255).astype(np.uint8)

    return denorm_img
```

I chose $\gamma$ = 0.45
c = 1

```python
image_gray = cv2.cvtColor(image_BGR, cv2.COLOR_BGR2GRAY)

new_image = powerLowTrans(image_gray, l=0.45, c=1)

plot_image(image_RGB, image_gray, title_1="Original image", title_2="Gray Image")
plot_image(image_RGB, new_image, title_1="Original image", title_2="Output Image")
```
